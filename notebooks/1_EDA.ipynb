# Exploratory Data Analysis (EDA) for Blood Cell Classification Data
In this section, we explore the dataset by visualizing sample images from each class and analyzing the class distribution.

```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import seaborn as sns
import random

data_dir = "../dataset"
class_names = os.listdir(data_dir)
```

```python
plt.figure(figsize=(15, 5))
for idx, class_name in enumerate(class_names):
    img_path = os.path.join(data_dir, class_name, os.listdir(os.path.join(data_dir, class_name))[0])
    img = Image.open(img_path)
    plt.subplot(1, 5, idx + 1)
    plt.imshow(img)
    plt.title(class_name,fontsize=16)
    plt.axis("off")
plt.suptitle("Sample Images Per Class", fontsize=18)
plt.tight_layout()
plt.show()
```

## Class Distribution
The dataset appears to be balanced, with each class containing a similar number of images. This balance reduces the risk of bias during model training and improves overall classification fairness.

```python
class_counts = {
    cls: len(os.listdir(os.path.join(data_dir, cls)))
    for cls in class_names
}

sns.barplot(
    x=list(class_counts.keys()),   
    y=list(class_counts.values())  
)

plt.title("Class Distribution")
plt.ylabel("Number of Images")
plt.show()
```

## Random Samples From Each Class
One random sample image from each blood cell class is displayed below.

```python
plt.figure(figsize=(15, 5))  # Genişliği fazla, çünkü yatay dizilecek

for idx, class_name in enumerate(class_names):
    class_path = os.path.join(data_dir, class_name)
    image_files = os.listdir(class_path)

    random_img = random.choice(image_files)
    img_path = os.path.join(class_path, random_img)
    img = Image.open(img_path)

    plt.subplot(1, len(class_names), idx + 1)
    plt.imshow(img)
    plt.title(class_name, fontsize=16)
    plt.axis("off")

plt.suptitle("One Random Image per Class", fontsize=18)
plt.tight_layout()
plt.show()
```
